```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib
import matplotlib.pyplot as plt
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_excel('URA.xlsx', header=0, parse_dates=[0])
```

```python
df.head()
```

```python
df['Date'].dtype
```

```python
df.index=df['Date']
```

```python
df.head()
```

```python
df['Adj Close'].fillna(value = df['Adj Close'].mean(), inplace=True)
```

```python
result = seasonal_decompose(df['Adj Close'], model='additive')
```

```python
result.plot()
```

```python
result2 = seasonal_decompose(df['Adj Close'], model='multiplicative')
```

```python
result2.plot()
```

```python
df['lag1']=df['Adj Close'].shift(1)
```

```python
df['diff_1'] = df['Adj Close'].diff(periods=1)
```

```python
df.head()
```

```python
result_b = seasonal_decompose(df.iloc[1:,2], model = 'additive')
result_b.plot()
```

```python
df['diff_2'] = df['diff_1'].diff(periods=2)
```

```python
df.head()
```

```python
df['diff_1'].plot()
```

```python
df['diff_2'].plot()
```

```python
df.index=df['Date']
result_c = seasonal_decompose(df.iloc[3:,4], model = 'additive')
result_c.plot()
```

```python
df['Adj Close'].plot()
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df['Adj Close'])
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
plot_pacf(df['Adj Close'], lags=15)
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = ARIMA(df['Adj Close'], order= (2,0,2))
```

```python
model_fit= model.fit()
```

```python
model_fit.summary()
```

```python
residuals = model_fit.resid
```

```python
residuals.plot()
```

```python
residuals.describe()
```

```python
output = model_fit.forecast()
```

```python
output
```

```python
model_fit.forecast(5)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df = pd.read_excel('URA.xlsx', header = 0, parse_dates = [0])
```

```python
model = SARIMAX(df['Adj Close'], order=(2,0,2), seasonal_order=(1,1,1,12))
```

```python
model_fit = model.fit()
```

```python
residuals = model_fit.resid
```

```python
residuals.plot()
```

```python
output = model_fit.forecast()
```

```python
output
```

```python
model_fit.forecast(50)
```

```python
yhat = model_fit.predict()
```

```python
yhat.head()
```

```python
yhat2 = model_fit.forecast(20)
```

```python
plt.plot(df['Adj Close'])
plt.plot(yhat, color = 'orange')
plt.plot(yhat2, color = 'red')
```
